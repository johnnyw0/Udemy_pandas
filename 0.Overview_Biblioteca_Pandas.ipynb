## **Análise de Dados com Pandas**

#### Bibliotecas do Python para análise de dados

Python é uma linguagem de programação muito popular dentro de Data Science e Analytics, pois dispomos de várias bibliotecas para esta finalidade.

As principais bibliotecas de análise de dados do Python são:

* NumPy: a biblioteca NumPy é fundamental para o universo de computação científica em Python. Com esta biblioteca podemos fazer operações entre matrizes, vetores, tensores, operações de álgebra linear, operações de cálculo numérico e diversas outras operações matemáticas.

* Pandas: é a principal opção para trabalhar com dados estruturados, pois fornece diversas funções para manipulação de dados, agrupamento, agregação, junção, seleção e diversas outras.

* Polars: biblioteca de processamento de dados projetada para ser rápida e escalável, possui sintaxe parecida com o Pandas em alguns procedimentos, suporta diversas funções para manipulação de dados.

* PySpark: versão do Apache Spark que suporta a linguagem Python, sendo que o Spark trabalha com processamento distribuído, o que possibilita trabalhar com grandes conjuntos de dados de forma eficiente e escalável. Fornece também funções para implementação de modelos de aprendizado de máquina.

No eixo de visualização de dados temos:

* Matplotlib: biblioteca de visualização de dados, permite criar gráficos como line plot, histogramas, scatter plot, bar plot, pie plot, dentre várias outras opções.

* Seaborn: biblioteca de visualização de dados baseada em Matplotlib, com foco em visualização estatística de dados. Com esta biblioteca conseguimos resultados mais atraentes.

Para visualizações dinâmicas temos opções como Plolty e Bokeh. Para construção de dashboards temos opções como Dash e Streamlit.

Já quanto ao uso do Python para Machine Learning e Estatística temos:

* Scikit-learn: biblioteca de aprendizado de máquina, fornece funções para as diversas etapas do flow de um projeto de Data Science, como pré-processamento de dados, seleção de features, avaliação de modelos, modelos de regressão e classificação, algoritmos de redução de dimensionalidade, dentre outras opções.

* SciPy: biblioteca focada em computação científica, fornece módulos para procedimentos estatísticos,de otimização, álgebra linear, dentre outras opções.

* Statsmodels: Siblioteca que fornece modelos estatísticos e funções para procedimentos estatísticos, como testes estatísticos, dentre várias outras opções.

#### **Overview da biblioteca Pandas**

* Estrutura de Dados do Pandas
    * O Pandas fornece duas estruturas de dados fundamentais: Series (coluna) e DataFrame (dado tabular).

* I/O
    * O Pandas possui integração com diversos formatos de arquivos de dados (csv, Excel, Json, SQL, parquet, etc.).
    
* Limpeza e Transformação de Dados
    * O Pandas possui diversas funções que possibilitam a preparação dos dados para  e limpeza dos dados. Tais operações englobam as seguintes tasks, dentre outras:
        * Identificação/Tratamento de Missing Values
        * Identificação/Tratamento de duplicidade
        * Renomeação e remoção de colunas
        * Transformação de dados
        
* Operações de Agregação
    * O Pandas possui diversas funções nativas para realizarmos operações de agregação (média, mediana, moda, desvio padrão, dentre outras estatísticas descritivas).
    
* Join e Concat
    * Às vezes, precisamos combinar dados de várias fontes. O Pandas oferece funções para fundir e concatenar conjuntos de dados, permitindo a criação de relacionamentos entre diferentes DataFrames.

* Análise Exploratória de Dados (EDA)
    * O Pandas é uma ferramenta essencial na etapa inicial da EDA, pois permite visualizar e resumir rapidamente os dados, fornecendo informações valiosas sobre a distribuição, tendências e padrões dos dados.

* Visualização de Dados
    * Embora o Pandas não seja uma biblioteca de visualização de dados, ele possui integração com outras bibliotecas populares, como Matplotlib e Seaborn, que permitem a criação de gráficos e visualizações impressionantes dos dados do Pandas.
